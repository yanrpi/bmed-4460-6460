```python
import numpy as np
import matplotlib.pyplot as plt

# https://scikit-image.org/
import skimage.filters as filters
import skimage.io as io
```

```python
image = io.imread('cells.tif')
print(type(image))
print(image.shape)
plt.imshow(image, cmap='gray')

thresh = filters.threshold_otsu(image)
print('Otsu threshold = {}'.format(thresh))
```

```python
image_binary = image > thresh
plt.imshow(image_binary, cmap='gray')
```

```python
#
# 3. Verify the computed threshold by plotting 
# the histogram and a vertical line to visually check
#
# hint: a histogram can be plotted directly by
#plt.hist(image, bins=256)
#
# hint: a vertical line can be plotted as
#plt.axvline(thresh, color='r')
#plt.show()
```

```python
hist, bins = np.histogram(image, bins=256, range=[0,256])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center')
plt.axvline(thresh, color='r')
plt.show()
```
