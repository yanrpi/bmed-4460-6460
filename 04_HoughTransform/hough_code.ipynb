```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((300, 300), dtype=np.uint8)

p0 = 50, 150
img[p0] = 255
p1 = 150, 250
img[p1] = 255
p2 = 250, 150
img[p2] = 255
p3 = 150, 50
img[p3] = 255

plt.imshow(img, cmap='gray')
plt.plot(p0[0], p0[1], 'w.')
plt.plot(p1[0], p1[1], 'w.')
plt.plot(p2[0], p2[1], 'w.')
plt.plot(p3[0], p3[1], 'w.')
```

```python
#
# Initialize Hough space
#
hough_space = np.zeros(img.shape, dtype=np.int32)
width, height = img.shape

# my guess of the radius
radius = 100

# Go through every pixel of the image
# This isn't an efficient approach.
# If you have idea to implement in an efficient way, that 
# will be great!

for w in range(width):
    for h in range(height):
        # check edge pixels
        if img[w,h] > 0:
            # Use np.mgrid may be more efficient, but efficiency
            # is not required as this moment.
            
            # Go through every pixel of the Hough space
            for i in range(width):
                for j in range(height):
                    dist = np.sqrt((i-w)**2 + (j-h)**2)
                    if dist > radius - 0.5 and dist < radius + 0.5:
                        hough_space[i, j] += 1
                        
plt.imshow(np.log(hough_space + 1), cmap='gray')
```
